```python
import pandas as pd
import matplotlib.pyplot as plt

# Load datasets
pop = pd.read_csv(r"C:\Users\Mayuri\Downloads\API_SP.POP.TOTL_DS2_en_csv_v2_76034\API_SP.POP.TOTL_DS2_en_csv_v2_76034.csv", skiprows=4)
male_pct = pd.read_csv(r"C:\Users\Mayuri\Downloads\API_SP.POP.TOTL.MA.ZS_DS2_en_csv_v2_7765\API_SP.POP.TOTL.MA.ZS_DS2_en_csv_v2_7765.csv", skiprows=4)
female_pct = pd.read_csv(r"C:\Users\Mayuri\Downloads\API_SP.POP.TOTL.FE.ZS_DS2_en_csv_v2_7756\API_SP.POP.TOTL.FE.ZS_DS2_en_csv_v2_7756.csv", skiprows=4)

# Focus on India
country = 'India'
india_pop = pop[pop['Country Name'] == country]

# Get years and population data
years = [str(year) for year in range(1960, 2024)]
population_values = india_pop[years].values.flatten()

# Plot as a bar chart
plt.figure(figsize=(16, 6))
plt.bar(years, population_values, color='skyblue')
plt.title('India Total Population (19602023)', fontsize=16)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Population', fontsize=12)
plt.xticks(rotation=90, fontsize=8)
plt.tight_layout()
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.show()

```
