```python
import pandas as pd
import matplotlib.pyplot as plt

# Realistic UN-based data for India, 2022
data = {
    'Gender': ['Male', 'Female'],
    'Population': [731_000_000, 686_000_000]
}

df = pd.DataFrame(data)

# Plotting the bar chart
plt.figure(figsize=(6, 5))
plt.bar(df['Gender'], df['Population'], color=['#4C72B0', '#DD8452'])
plt.title('India Population by Gender — 2022', fontsize=14)
plt.xlabel('Gender')
plt.ylabel('Population (Number of People)')
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()

```
