```python
import pandas as pd
import matplotlib.pyplot as plt

# Load datasets
pop = pd.read_csv(r"C:\Users\Mayuri\Downloads\API_SP.POP.TOTL_DS2_en_csv_v2_76034\API_SP.POP.TOTL_DS2_en_csv_v2_76034.csv", skiprows=4)
male_pct = pd.read_csv(r"C:\Users\Mayuri\Downloads\API_SP.POP.TOTL.MA.ZS_DS2_en_csv_v2_7765\API_SP.POP.TOTL.MA.ZS_DS2_en_csv_v2_7765.csv", skiprows=4)
female_pct = pd.read_csv(r"C:\Users\Mayuri\Downloads\API_SP.POP.TOTL.FE.ZS_DS2_en_csv_v2_7756\API_SP.POP.TOTL.FE.ZS_DS2_en_csv_v2_7756.csv", skiprows=4)

# Focus on India
country = 'India'
years = [str(year) for year in range(1960, 2024)]

# Total population
india_pop = pop[pop['Country Name'] == country][years].values.flatten()
# Male and female % per year
male_percent = male_pct[male_pct['Country Name'] == country][years].values.flatten()
female_percent = female_pct[female_pct['Country Name'] == country][years].values.flatten()

# Compute actual male and female population
male_population = (male_percent / 100) * india_pop
female_population = (female_percent / 100) * india_pop

# Plot stacked bar chart
plt.figure(figsize=(18, 6))
plt.bar(years, male_population, label='Male', color='blue')
plt.bar(years, female_population, bottom=male_population, label='Female', color='pink')
plt.title('India Gender-wise Population Distribution (19602023)', fontsize=16)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Population', fontsize=12)
plt.xticks(rotation=90, fontsize=8)
plt.legend()
plt.tight_layout()
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.show()


```
